```python
import geopandas as gpd
import pandas as pd
import numpy as np
```

```python
adjusted_pop = gpd.read_file("adjusted_population.shp")
prison_capacity = gpd.read_file("ny_state_prisons_capacity.csv")
correctional_facilities = gpd.read_file("correctional_facility_locations.shp")
```

```python
adjusted_pop['name'] = adjusted_pop['name'].str.replace(', New York', '')
```

```python
adjusted_pop
```

```python
counties = adjusted_pop[['name', 'geometry']]
```

```python
counties
```

```python
prison_capacity
```

```python
correctional_facilities
```

```python
prisons = correctional_facilities.merge(prison_capacity, left_on='name', right_on='Name')
```

```python
prisons['Capacity']= prisons['Capacity'].replace(r'^\s*$', np.nan, regex=True)
```

```python
prisons['Capacity']= prisons['Capacity'].fillna(0)
```

```python
prisons['Capacity'] = prisons['Capacity'].astype(float)
```

```python
prisons = gpd.GeoDataFrame(prisons, geometry='geometry_x')
```

```python
counties.crs
```

```python
counties = counties.to_crs(4326)
```

```python
prisons.crs
```

```python
prisoners_per_county = counties.sjoin(prisons, how='left')
```

```python
prisoners_per_county
```

```python
prisoners_per_county_agg = prisoners_per_county.groupby('name_left')['Capacity'].sum()
```

```python
pd.DataFrame(prisoners_per_county_agg)
```

```python
adjusted_pop
```

```python
adjusted_pop_w_prisoners = adjusted_pop.merge(prisoners_per_county_agg, left_on='name', right_on='name_left')
```

```python
adjusted_pop_w_prisoners[adjusted_pop_w_prisoners['name']=='Kings County']
```

```python
adjusted_pop_w_prisoners['ttl_pp_prisoners'] = adjusted_pop_w_prisoners['ttl_pp_'] + adjusted_pop_w_prisoners['Capacity'] 
```

```python
adjusted_pop_w_prisoners[adjusted_pop_w_prisoners['name']=='New York County']
```

```python
adjusted_pop_w_prisoners['ttl_pp_prisoners'] = adjusted_pop_w_prisoners['ttl_pp_prisoners'].clip(lower=0)
```

```python
adjusted_pop_w_prisoners
```

```python
adjusted_pop_w_prisoners.to_file("adjusted_pop.shp")
```
