```python
import geopandas as gpd
import pandas as pd
```

```python
employment = gpd.read_file("employment.shp")
```

```python
index = employment[['cnty_nm', 'geometry']]
```

```python
index = index.drop_duplicates()
```

```python
index
```

```python
employment
```

```python
manufacturing_p = employment.pivot(index=['cnty_nm'], values='mnfctrng_p', columns = 'year').reset_index()
```

```python
manufacturing_p = manufacturing_p.rename(columns = {'1970.0':'1970_m', '1980.0':'1980_m', '1990.0':'1990_m', '2000.0':'2000_m', '2010.0':'2010_m'})
```

```python
manufacturing_p.columns
```

```python
manufacturing_p.columns = ['county', '1970', '1980', '1990', '2000', '2010']
```

```python
manufacturing_p
```

```python
pso_p = employment.pivot(index=['cnty_nm'], values='pso_prc', columns = 'year').reset_index()
```

```python
pso_p.columns = ['county', '1970', '1980', '1990', '2000', '2010']
```

```python
pso_p
```

```python
index_manufacturing = index.merge(manufacturing_p, left_on='cnty_nm', right_on='county')
```

```python
index_manufacturing
```

```python
wide_data = index_manufacturing.merge(pso_p, left_on='cnty_nm', right_on='county')
```

```python
wide_data
```

```python
wide_data.to_file("employment_wide.shp")
```
